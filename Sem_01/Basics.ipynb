# Let's start on the basics. Variables and arrays

### First, import some libraries

```
import numpy as np
```

```
N = 10
array_1 = np.arange(N)# Create an array of N values
array_2 = np.random.randint(low= 0, high = 10, size=N) # Create an array of random int values
            # with the same lenght as array_1

print(array_1, array_2) # print both array_1 and array_2
print(array_1.shape, array_2.shape) # print both array_1 and array_2 shapes
print(np.sum(array_1), np.sum(array_2))# print summation of array_1 and array_2
```

```
array_1[2] = 70 # Replace the value of array_1 in pos 3 by 70
print(array_1)# Print summation again
print(np.sum(array_1))
```

```
result_array = array_1*array_2 # Apply element-wise multiplication among array_1 and array_2
print("result element wise: ", result_array) # Print results
```

### Going further

```
# Messing everything around.
img = cv2.imread( "/content/lena.jpg", 0 )
plt.imshow(img)
```

```
# Playing with image intensities
img_gray = cv2.imread("/content/lena.jpg")
alto, ancho, canales = img_gray.shape

for r in range(alto)  :  # Go through rows
  for c in range(ancho) : # Go through cols
    if img_gray[r, c, 0] > 127 : # Create a simple condition
      img_gray[r, c, 0]+= 100 # Do somethin
      img_gray[r, c, 1]+= 500

plt.imshow(img_gray)
```
